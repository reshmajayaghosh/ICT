# **2. Weather Data Analyzer**

a. Analyze weather data (temperature, humidity, etc.) for a given region

b. Use variables and data types to store weather data


```python

temperature = [30,32,31,29,33]
humidity = [65,70,68,60,72]

```


c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)

```python
current_temperature = 32
current_humidity = 70

if current_temperature > 30 and current_humidity < 60:
    print("Weather is Sunny")
elif current_humidity >= 60:
    print("Weather is Rainy")
else:
    print("Weather is Pleasant")
```


d. Use functions to calculate average temperature and humidity, and print
results


```python
def average_temperature(temp_list):
    avg_temp = sum(temp_list) / len(temp_list)
    print("Average Temperature:", avg_temp)


def average_humidity(humidity_list):
    avg_hum = sum(humidity_list) / len(humidity_list)
    print("Average Humidity:", avg_hum)

```

```python
average_temperature(temperature)
average_humidity(humidity)
```
