# **NumPy Problems**

1. **Basic Array Operations**


**Convert the mpg column into a NumPy array and calculate:**

4. **Array Manipulation**

Convert the acceleration column into a NumPy array and normalize its values (scale between 0 and 1).

```python
acc_array = df['acceleration'].to_numpy()

acc_min = np.min(acc_array)
acc_max = np.max(acc_array)

acc_normalized = (acc_array - acc_min) / (acc_max - acc_min)
acc_normalized
```

5. **Broadcasting**

Increase all horsepower values by 10% and store the updated values in a new NumPy array. Handle missing data (if any) by replacing it with the mean of the column before applying the increase.

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
horsepower_array = df['horsepower'].to_numpy()

# Replace missing values with mean
hp_mean = np.nanmean(horsepower_array)
horsepower_array = np.where(np.isnan(horsepower_array), hp_mean, horsepower_array)

# Increase by 10%
horsepower_updated = horsepower_array * 1.10
horsepower_updated
```

10. **Finding Anomalies**

Identify potential outliers in the mpg column using the Interquartile Range (IQR) method. Specifically:


â— Calculate the IQR for mpg.

â— Defi ne outliers as values less than Q1 - 1.5 * IQR or greater than Q3 + 1.5 * IQR.

â— Create a DataFrame of cars classifi ed as outliers, displaying car_name, mpg, and model_year.

```python
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_df = df[
    (df['mpg'] < lower_bound) |
    (df['mpg'] > upper_bound)
][['car name', 'mpg', 'model year']]

outliers_df
```
